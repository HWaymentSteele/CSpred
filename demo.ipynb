# UCBShift Demo

This notebook demonstrates how to use UCBShift to predict NMR chemical shifts from a PDB file.

```python
from cspred import predict_shifts
```

## Basic usage — predict all atom types

By default, `predict_shifts` predicts shifts for all backbone atoms (H, HA, C, CA, CB, N) using both the UCBShift-X (ML) and UCBShift-Y (transfer prediction) modules.

```python
pdb_id = "1UBQ"
df = predict_shifts(f"{pdb_id}_hyd.pdb", pH=7.0)
df.head()
```

## ML-only prediction (works in Colab)

Set `TP=False` to skip UCBShift-Y, which requires BLAST and mTM-align. ML-only predictions run without any external tools.

```python
df_ml = predict_shifts(f"{pdb_id}_hyd.pdb", pH=7.0, TP=False)
df_ml.head()
```

## Predict a subset of atom types

Pass `atom_types` to restrict prediction to specific atoms. This avoids loading unused models and speeds up the calculation.

```python
df_subset = predict_shifts(f"{pdb_id}_hyd.pdb", pH=7.0, TP=False, atom_types=["H", "N", "CA"])
df_subset.head()
```

## Available atom types

| Atom | Description |
|------|-------------|
| `H`  | Backbone amide proton |
| `HA` | Alpha proton |
| `C`  | Carbonyl carbon |
| `CA` | Alpha carbon |
| `CB` | Beta carbon |
| `N`  | Backbone amide nitrogen |

## Output columns

| Column | Description |
|--------|-------------|
| `RESNUM` | Residue number |
| `RESNAME` | Residue name (3-letter code) |
| `[atom]_X` | UCBShift-X (ML) prediction |
| `[atom]_Y` | UCBShift-Y (transfer prediction) |
| `[atom]_UCBShift` | Final combined prediction (when both modules are used) |

## `predict_shifts` parameters

| Parameter | Type | Default | Description |
|-----------|------|---------|-------------|
| `pdb_file_name` | str | — | Path to the (hydrogenated) PDB file |
| `pH` | float | 5.0 | pH for prediction |
| `TP` | bool | True | Use UCBShift-Y transfer prediction module |
| `ML` | bool | True | Use UCBShift-X machine learning module |
| `test` | bool | False | Exclude mode for UCBShift-Y (for benchmarking) |
| `atom_types` | list of str | None (all) | Atoms to predict — any subset of `["H","HA","C","CA","CB","N"]` |
