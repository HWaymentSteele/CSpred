```python
#@title Install packages and download UCBshift weights

# Install packages
!pip install git+https://github.com/HWaymentSteele/CSpred.git hydride

# # Download and extract model weights
from huggingface_hub import snapshot_download
from cspred import predict_shifts
import os
import subprocess as sp

repo_id = "hkws/ucbshift_weights"
local_dir = "./ucbshift_weights"

if not os.path.exists(f'{local_dir}/CA_R0.sav'):
  os.makedirs(local_dir, exist_ok=True)
  snapshot_download(repo_id=repo_id, repo_type="model", local_dir=local_dir)
```

```python
protein='1UBQ'

#fetch protein
sp.run(["wget", f"https://files.rcsb.org/download/{protein}.pdb"], check=True)

#add hydrogens with hydride package
sp.run(["hydride", "-i", f"{protein}.pdb", "-o", f"{protein}_hyd.pdb"], check=True)

#to specify only doing subset of atoms, otherwise omit
import toolbox
toolbox.ATOMS = ['H','N']

#generate predictions
df = predict_shifts(f"{protein}_hyd.pdb", pH=7.0, TP=False)
```
